# Data cleaning

```python
print(csv_df['Item Fat Content'].unique())
```

```python
# Creating a copy of the DataFrame to avoid modifying the original data
csv_df_copy = csv_df.copy()
```

```python
# Standardizing the 'Item Fat Content' column values
csv_df_copy['Item Fat Content'] = csv_df_copy['Item Fat Content'].replace({'LF': 'Low Fat', 'low fat': 'Low Fat', 'reg': 'Regular', 'Regular ': 'Regular'})

print(csv_df_copy['Item Fat Content'].unique())
```

## **Business requirements**

### **KPI's requirements**

```python
# Total sales

total_sales = csv_df_copy['Sales'].sum()
print(f'Total sales: ${total_sales:,.2f}')
```

```python
# Average sales

average_sales = csv_df_copy['Sales'].mean()
print(f'Average sales: ${average_sales:,.2f}')
```

```python
# Number of items

number_of_items_sold = csv_df_copy['Sales'].count()
print(f'Number of items sold: {number_of_items_sold:,.2f}')
```

```python
# Average ratings

average_ratings = csv_df_copy['Rating'].mean()
print(f"Average ratings: {average_ratings:,.2f}")
```

### **Charts requirements**

#### **Total Sales by Fat Content**

```python
sales_by_fat = csv_df_copy.groupby('Item Fat Content')['Sales'].sum()

sales_by_fat
```

```python
plt.pie(sales_by_fat, labels = sales_by_fat.index, autopct='%.1f%%', startangle=90)
plt.title('Total Sales by Fat Content')
plt.axis('equal')
plt
```

#### **Total Sales by Item Type**

```python
sales_by_type = csv_df_copy.groupby('Item Type')['Sales'].sum().sort_values(ascending=False)

plt.figure(figsize=(10, 6))
bars = plt.bar(sales_by_type.index, sales_by_type.values, color=sns.color_palette("viridis", len(sales_by_type)))

plt.xticks(rotation=-45, ha='center')
plt.xlabel('Item Type')
plt.ylabel('Total Sales ($)')
plt.title('Total Sales by Item Type')

for bar in bars:
        plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), f'${bar.get_height():,.0f}', ha='center', va='bottom', fontsize=8)

plt.tight_layout()
plt.show()
```

#### **Fat Content by Outlet for Total Sales**

```python
grouped = csv_df_copy.groupby(['Outlet Location Type', 'Item Fat Content'])['Sales'].sum().unstack()
grouped = grouped[['Low Fat', 'Regular']]

ax = grouped.plot(kind='bar', title='Sales by Outlet Location Type and Fat Content', figsize=(8, 5), color=["#1e42e2", "#01bedb"])
plt.xlabel('Outlet Location Tier')
plt.ylabel('Total Sales ($)')
plt.legend(title='Item Fat Content')
plt.tight_layout()
plt.show()
```

#### **Total Sales by Outlet Establishment**

```python
sales_by_year = csv_df_copy.groupby('Outlet Establishment Year')['Sales'].sum().sort_index()

plt.figure(figsize=(9, 5))
plt.plot(sales_by_year.index, sales_by_year.values, marker='o', linestyle='-')

plt.xlabel('Outlet Establishment Year')
plt.ylabel('Total Sales')
plt.title('Outlet Establishment')

for x, y in zip(sales_by_year.index, sales_by_year.values):
    plt.text(x, y, f'{y:.0f}', ha='center', va='bottom', fontsize=8)

plt.tight_layout()
plt.show()
```

#### **Sales by Outlet Size**

```python
sales_by_size = csv_df_copy.groupby('Outlet Size')['Sales'].sum()

plt.figure(figsize=(4, 4))
plt.pie(
    sales_by_size, 
    labels=sales_by_size.index, 
    autopct='%1.1f%%', 
    startangle=90
)
plt.title('Outlet Size')
plt.tight_layout()
plt.show()
```

#### **Sales by Outlet Location**

```python
sales_by_location = csv_df_copy.groupby('Outlet Location Type')['Sales'].sum().reset_index()
sales_by_location = sales_by_location.sort_values('Sales', ascending=False)

plt.figure(figsize=(8, 4)) # Smaller height, enough width
ax = sns.barplot(x='Sales', y='Outlet Location Type', data=sales_by_location)

plt.title('Total Sales by Outlet Location Type')
plt.xlabel('Total Sales')
plt.ylabel('Outlet Location Type')

plt.tight_layout() # Ensures layout fits without scroll
plt.show()
```
